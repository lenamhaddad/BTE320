```python
def calcFee(hour):
  fee = 5 + 2.5 * hour
  if fee < 10:
    fee = 10.0
  elif fee > 20:
    fee = 20.0
  print(fee)

def main():
  hour = int(input('Enter number of hours parked: '))
  calcFee(hour)

main()
```
