```python
d = {
    "P-01": 9.95,
    "P-02": 10.50
}

product_id = input("Enter product ID: ")

if product_id == "P-01":
  print(d.value())
elif product_id == "P-02":
    print(d.value())
else:
  print("Product ID Not Found")
```

```python
products = {
    "P-01": 9.95,
    "P-02": 10.5
}

id = input("Enter product ID: ")

if id in products:
  print(products[id])
else:
  print(f"Product {id} not found...")
```

```python
products = {
    "P-01": 9.95,
    "P-02": 10.5
}

id = input("Enter product ID: ")

print(products.get(id, f"Product {id} not found..."))
```
