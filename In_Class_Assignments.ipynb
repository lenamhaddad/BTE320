# Income Statement

```python

sales = float(input("Enter sales: $"))
rate = float(input("Enter tax rate (%): "))
cogs = 0.4 * sales
gross_profit = sales - cogs
taxes = rate * gross_profit
net_income = gross_profit - taxes
print(f"\nIncome Statement")
print(f"sales: {sales}$")
print(f"rate: {rate}%")
print(f"cogs: {cogs}$")
print(f"gross profit: {gross_profit}$")
print(f"taxes: {taxes}$")
print(f"net income: {net_income}$")
```
