````python
grade = int(input("Enter a grade (0-100): "))

if grade >= 93:
  print("A")
elif grade >= 97:
  print("A+")
elif grade >= 90:
  print("A-")
````

````python
email = "URGENT!! Click on the link: www.phis.org and receive a reward!"
has_urgent = "URGENT"
has_link = "www."

if has_urgent in email:
  if has_link in email:
    print("Phishing!")
  else:
    print("Spam!")
else:
  print("Legit!")
````

````python
email = "URGENT!! Click on the link: www.phis.org and receive a reward!"
has_urgent = "URGENT"
has_link = "www."

if has_urgent in email and has_link in email:
  print("Phishing!")
elif has_urgent in email:
  print("Spam!")
else:
  print("Legit!")
````
